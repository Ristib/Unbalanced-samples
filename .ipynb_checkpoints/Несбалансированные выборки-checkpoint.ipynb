Разделить датасет на обучающую и тестовую выборки в пропорции 80:20. Разделение должно быть стратифицированным по таргету Risk.

В этом и следующих заданиях использую random_state = 1.

```python
from sklearn.model_selection import train_test_split

# Разделение данных на признаки и целевую переменную
X = data.drop('Risk', axis=1)
y = y = data['Risk'].apply(lambda x: 1 if x == 'good' else 0)

# Разделение данных на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=1)

# Проверка распределения целевой переменной в обучающей и тестовой выборках
print("Train target distribution:\n", y_train.value_counts(normalize=True))
print("\nTest target distribution:\n", y_test.value_counts(normalize=True))
```

Провожу этап очистки и подготовки данных (data preparation) и подготовка данные к моделированию.

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Выделение числовых и категориальных колонок
numeric_features = ['Age', 'Credit amount', 'Duration']
categorical_features = [col for col in X.columns if col not in numeric_features]

# Создание трансформеров для числовых и категориальных данных
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_transformer_most_frequent = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(drop='first'))
])

# Создание трансформеров для категориальных данных (с заполнением пропусков значением 'missing' и one-hot кодированием)
categorical_transformer_missing = Pipeline(steps=[
    ('imputer', SimpleImputer(fill_value='missing', strategy='constant')),
    ('onehot', OneHotEncoder(drop='first'))
])    

# Создание трансформеров для категориальных данных (с заполнением пропусков значением 'unknown' и one-hot кодированием)
categorical_transformer_unknown = Pipeline(steps=[
    ('imputer', SimpleImputer(fill_value='unknown', strategy='constant')),
    ('onehot', OneHotEncoder(drop='first'))
])

categorical_transformer = categorical_transformer_most_frequent

# Объединение трансформеров в единый ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Применение преобразований к обучающей выборке
X_train_prepared = preprocessor.fit_transform(X_train)
X_test_prepared = preprocessor.transform(X_test)

# Проверка размеров преобразованных данных
print("Размеры обучающей выборки:", X_train_prepared.shape)
print("Размеры тестовой выборки:", X_test_prepared.shape)
```

Сбалансирую обучающую выборку методом SMOTE и обучу модель RandomForestClassifier. Подберу оптимальные гиперпараметры с помощью GridSearch. Посчитаю метрику ROC-AUC на тестовой выборке.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline

# Создание пайплайна с SMOTE и RandomForest
pipeline = ImbPipeline(steps=[
    ('preprocessor', preprocessor),
    ('smote', SMOTE(random_state=1)),
    ('classifier', RandomForestClassifier(random_state=1))
])

# Определение параметров для GridSearch
param_grid = {
    'classifier__n_estimators': [100, 200],
    'classifier__max_depth': [None, 10, 20],
    'classifier__min_samples_split': [2, 5],
    'classifier__min_samples_leaf': [1, 2]
}

# Создание GridSearchCV
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='roc_auc', n_jobs=-1)

# Обучение модели с подбором гиперпараметров
grid_search.fit(X_train, y_train)

# Вывод лучших параметров
print("Best parameters found: ", grid_search.best_params_)
```

### Оценка модели на тестовой выборке

```python
# Предсказания вероятностей на тестовой выборке
y_pred_proba = grid_search.predict_proba(X_test)[:, 1]

# Вычисление метрики ROC-AUC
roc_auc = roc_auc_score(y_test, y_pred_proba)
print("ROC-AUC on the test set: ", roc_auc)
```

Обучение модели RandomForestClassifier с использованием весов классов. Подберу оптимальные веса и гиперпараметры с помощью GridSearch. Посчитаю метрику ROC-AUC на тестовой выборке.

```python
# Создание пайплайна с RandomForest
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=1))
])

# Определение параметров для GridSearch
param_grid = {
    'classifier__n_estimators': [100, 200],
    'classifier__max_depth': [None, 10, 20],
    'classifier__min_samples_split': [2, 5],
    'classifier__min_samples_leaf': [1, 2],
    'classifier__class_weight': [None, 'balanced', 'balanced_subsample'] + [{0: 1, 1: w} for w in [1, 2, 5, 10]]
}

# Создание GridSearchCV
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='roc_auc', n_jobs=-1)

# Обучение модели с подбором гиперпараметров
grid_search.fit(X_train, y_train)

# Вывод лучших параметров
print("Best parameters found: ", grid_search.best_params_)
```

### Оценка модели на тестовой выборке

```python
# Предсказания вероятностей на тестовой выборке
y_pred_proba = grid_search.predict_proba(X_test)[:, 1]

# Вычисление метрики ROC-AUC
roc_auc = roc_auc_score(y_test, y_pred_proba)
print("ROC-AUC on the test set: ", roc_auc)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score
from sklearn.model_selection import cross_val_score
import numpy as np

X_train['Saving accounts'].fillna('no_info', inplace=True)
X_train['Checking account'].fillna('no_info', inplace=True)
X_test['Saving accounts'].fillna('no_info', inplace=True)
X_test['Checking account'].fillna('no_info', inplace=True)

categorical_transformer_no_info = Pipeline(steps=[
    ('onehot', OneHotEncoder(drop='first', handle_unknown='ignore'))
])

# Функция для создания и оценки модели
def evaluate_model(categorical_transformer):
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', numeric_transformer, numeric_features),
            ('cat', categorical_transformer, categorical_features)
        ])
    
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', RandomForestClassifier(random_state=1))
    ])
    
    # Обучение модели
    pipeline.fit(X_train, y_train)
    
    # Предсказания
    y_pred = pipeline.predict(X_test)
    y_pred_proba = pipeline.predict_proba(X_test)[:, 1]
    
    # Вычисление метрик
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred_proba)
    
    return accuracy, precision, recall, roc_auc

# Оценка моделей
results = {}
strategies = {
    'most_frequent': categorical_transformer_most_frequent,
    'missing': categorical_transformer_missing,
    'unknown': categorical_transformer_unknown,
    'no_info': categorical_transformer_no_info
}

for name, transformer in strategies.items():
    results[name] = evaluate_model(transformer)

# Вывод результатов
for name, metrics in results.items():
    print(f"Strategy: {name}")
    print(f"  Accuracy: {metrics[0]:.4f}")
    print(f"  Precision: {metrics[1]:.4f}")
    print(f"  Recall: {metrics[2]:.4f}")
    print(f"  ROC AUC: {metrics[3]:.4f}\n")
```
